# üí≥ Credit Card Fraud Detection Using Unsupervised Learning 

## üö© Problem Description

In this project, we aim to detect potentially fraudulent credit card transactions using unsupervised learning methods ‚Äî specifically PCA for dimensionality reduction and KMeans for clustering.

Rather than relying on manually assigned labels ‚Äî which are often error-prone, delayed, or biased ‚Äî our approach looks for anomalous patterns in the data itself. This avoids dependence on heuristics like repetition, device fingerprinting, or geographic stereotypes, which can introduce bias, frustrate legitimate customers, and unnecessarily disrupt commerce.

Since labeled fraud data is limited or unavailable during training, the goal is to identify suspicious transactions based on behavioral outliers. By grouping similar transactions and analyzing small or distinct clusters, we aim to surface patterns that may correlate with fraud without requiring labeled data.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

## üîç Exploratory Data Analysis (EDA)

First, let's check the dataset's structure and look at basic statistics.

```python
# Basic info
df.info()
```

```python
# Summary statistics
df.describe()
```

```python
# Check the class distribution (fraud vs non-fraud)
sns.countplot(x='Class', data=df)
plt.title('Class Distribution (0 = Legit, 1 = Fraud)')
plt.show()
```

The dataset is highly imbalanced, with very few fraud cases. This supports using unsupervised methods to detect anomalies, rather than relying solely on supervised classifiers.

```python
# Scale the features (exclude the 'Class' column)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df.drop('Class', axis=1))
```

## Dimensionality Reduction Using PCA

PCA reduces dimensionality while preserving variance, making clustering more effective.

```python
pca = PCA(n_components=0.95)  # Keep 95% variance
X_pca = pca.fit_transform(X_scaled)

print(f'Original number of features: {X_scaled.shape[1]}')
print(f'Reduced number of features after PCA: {X_pca.shape[1]}')
```

## K-Means Clustering

We apply KMeans on PCA-reduced data to group similar transactions. We expect fraudulent ones to form distinct or smaller clusters.


```python
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(X_pca)
```

```python
# Add cluster labels to the dataframe
df['cluster'] = clusters
```

## Results and Evaluation

Let's examine the size of each cluster and how fraud cases are distributed.

```python
# Cluster sizes
print(df['cluster'].value_counts())
```

```python
# Cross-tabulation of clusters vs fraud class
print(pd.crosstab(df['cluster'], df['Class']))
```

Cluster 2 contains a disproportionately higher number of fraudulent transactions, indicating the model may be grouping anomalies together.
## Visualization of Clusters in PCA Space

```python
plt.figure(figsize=(10,6))
plt.scatter(X_pca[:,0], X_pca[:,1], c=clusters, cmap='viridis', s=5)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('K-means Clusters on PCA-Reduced Data')
plt.show()
```

### ‚ö†Ô∏è Adding Context: Where Are the Fraudulent Transactions?

Now that we've visualized the clusters formed using KMeans on PCA-reduced data, our next step is to evaluate how fraud cases are distributed across those clusters.

By overlaying the known fraud labels, we can assess whether any clusters are more strongly associated with fraudulent activity ‚Äî a valuable insight even in this unsupervised setting.

```python
# Add PCA components to the original DataFrame for plotting
df['PCA1'] = X_pca[:, 0]
df['PCA2'] = X_pca[:, 1]

# Plot clusters and highlight known frauds in red
plt.figure(figsize=(10, 6))
plt.scatter(df['PCA1'], df['PCA2'], c=clusters, cmap='viridis', s=3, alpha=0.5)
plt.scatter(df[df['Class'] == 1]['PCA1'], df[df['Class'] == 1]['PCA2'],
            color='red', s=8, label='Fraud', alpha=0.9)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('KMeans Clusters with Highlighted Fraud Cases')
plt.legend()
plt.show()
```

### üìå Discussion and Conclusion

**PCA** successfully reduced the feature space, capturing most of the variance with fewer components, which simplified the clustering process and enabled visualization.

**K-means** clustering identified distinct transaction groups, and after overlaying known fraud cases, we observed that one cluster (visually shown in purple) contained a notably higher proportion of fraud. This alignment suggests that, despite no label guidance, KMeans captured patterns indicative of fraudulent behavior.

While this unsupervised approach cannot perfectly separate fraud cases, it demonstrates value as a lightweight anomaly detection tool, especially in scenarios where labeled data is incomplete, unavailable, or potentially biased.
